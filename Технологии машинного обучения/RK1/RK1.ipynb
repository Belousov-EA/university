```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../data/toy_dataset.csv')
```

```python
# Проверим наличие пустых значений
for col in data.columns:
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True, fmt='.3f')
```

## Вывод по корреляционному анализу

Исходя из кореляционного анализа нельзя сделать вывод о зависимости дохода от возраста или номера в таблице, соответственно, эти данные ненужно использовать для построения линейных моделей. К остальным атрибутам, имеющим тип object, нельзя применить корреляционный анализ.

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='Age', y='Income', data=data)
```

как и ожидалось, корреляции между данными не наблюдается

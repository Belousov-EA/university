```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
%matplotlib inline
```

```python
# 171 + 22 x + 3 y >= 0
def f1(x):
    return float(171 + 22*x)/-3.

# 130 - 5 x + 10 y >= 0
def f2(x):
    return float(130 - 5*x)/-10.

# 95 + 7 x + 9 y >= 0
def f3(x):
    return float(95 + 7*x)/-9.

# 44 + 6 x - 4 y >= 0
def f4(x):
    return float(44 + 6*x)/4.

# 292 - 19 x - 22 y >= 0
def f5(x):
    return float(292 - 19*x)/22.

# -13x - 4y
def F(x):
    return float(-13*x)/4.
```

```python
#171 + 22 x + 3 y >= 0 и 95 + 7 x + 9 y >= 0
M1 = np.array([[22., 3.], [7., 9.]])
v1 = np.array([-171, -95])
t1 = np.linalg.solve(M1, v1)

#130 - 5 x + 10 y >= 0 и 95 + 7 x + 9 y >= 0
M2 = np.array([[-5., 10.], [7., 9.]])
v2 = np.array([-130, -95])
t2 = np.linalg.solve(M2, v2)

#130 - 5 x + 10 y >= 0 и 292 - 19 x - 22 y >= 0
M3 = np.array([[-5., 10.], [-19., -22.]])
v3 = np.array([-130, -292])
t3 = np.linalg.solve(M3, v3)

# 44 + 6 x - 4 y >= 0 и 292 - 19 x - 22 y >= 0
M4 = np.array([[6., -4.], [-19., -22.]])
v4 = np.array([-44, -292])
t4 = np.linalg.solve(M4, v4)

# 44 + 6 x - 4 y >= 0 и 171 + 22 x + 3 y >= 0
M5 = np.array([[6., -4.], [22., 3.]])
v5 = np.array([-44, -171])
t5 = np.linalg.solve(M5, v5)

# 130 - 5 x + 10 y >= 0 и 171 + 22 x + 3 y >= 0
M = np.array([[6., -4.], [22., 3.]])
v = np.array([-44, -171])
t = np.linalg.solve(M5, v5)
```

```python
def Fu(t):
    return float(-13*t[0] - 4*t[1])
index = -1
min = float('+inf')
count = 0
for i in [t1, t2, t3, t4, t5]:
    count += 1
    print( count,' ', i, ' ', Fu(i))
    if min > Fu(i):
        min = Fu(i)
        index = count
index
```

```python
x = [-10, 20]
plt.figure(figsize=(15,15))
plt.plot([-10, 0], list(map(f1, [-10, 0])), 'b', label = '171 + 22 x + 3 y >= 0')
plt.plot(x, list(map(f2, x)), 'g', label = '130 - 5 x + 10 y >= 0')
plt.plot(x, list(map(f3, x)), 'r', label = '95 + 7 x + 9 y >= 0')
plt.plot(x, list(map(f4, x)), 'c', label = '44 + 6 x - 4 y >= 0')
plt.plot(x, list(map(f5, x)), 'm', label = '292 - 19 x - 22 y >= 0')
plt.plot([t1[0], t2[0]], [t1[1], t2[1]], 'k')
plt.plot([t2[0], t3[0]], [t2[1], t3[1]], 'k')
plt.plot([t3[0], t4[0]], [t3[1], t4[1]], 'k')
plt.plot([t4[0], t5[0]], [t4[1], t5[1]], 'k')
plt.plot([t5[0], t1[0]], [t5[1], t1[1]], 'k')
plt.plot(x, list(map(F, x)), 'y', label = 'f(x, y)')
#plt.plot(t1[0], t1[1], 'ok')
#plt.plot(t2[0], t2[1], 'ok')
#plt.plot(t3[0], t3[1], 'ok')
#plt.plot(t4[0], t4[1], 'ok')
#plt.plot(t5[0], t5[1], 'ok')

plt.plot(t3[0], t3[1], 'ok', label = 'min')
plt.grid()
plt.legend()
```
